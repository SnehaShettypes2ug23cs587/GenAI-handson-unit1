```python
sentiment_analyzer = pipeline(
    "sentiment-analysis",
    model="distilbert-base-uncased-finetuned-sst-2-english"
)
```

```python
review = "The app interface is confusing, but the features are very powerful."

sentiment_analyzer(review)
```

```python
reviews = [
    "The service was amazing and very fast.",
    "The product stopped working after two days.",
    "Customer support was helpful and polite."
]

sentiment_analyzer(reviews)
```
